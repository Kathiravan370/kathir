```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("dataset_folder.csv", encoding='latin1')
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
product_sales = df.groupby('Description')['TotalSales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
plt.bar(product_sales.index, product_sales.values)
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Top 10 Products by Sales")
plt.xticks(rotation=90)
plt.show()
plt.figure(figsize=(10,5))
plt.plot(product_sales.index, product_sales.values, marker='o')
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Sales Trend of Top 10 Products")
plt.xticks(rotation=90)
plt.show()
```
