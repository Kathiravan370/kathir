```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("Housing.csv")
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())
numeric_df = df.select_dtypes(include=np.number)
corr_matrix = numeric_df.corr()
plt.figure(figsize=(8,6))
plt.imshow(corr_matrix)
plt.colorbar()
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title("Correlation Heatmap")
plt.show()
if 'price' in numeric_df.columns and 'area' in numeric_df.columns:
    plt.figure(figsize=(6,4))
    plt.scatter(numeric_df['area'], numeric_df['price'])
    plt.xlabel("Area")
    plt.ylabel("Price")
    plt.title("Area vs Price")
    plt.show()
if 'price' in numeric_df.columns and 'bedrooms' in numeric_df.columns:
    plt.figure(figsize=(6,4))
    plt.scatter(numeric_df['bedrooms'], numeric_df['price'])
    plt.xlabel("Bedrooms")
    plt.ylabel("Price")
    plt.title("Bedrooms vs Price")
    plt.show()
print("Average Price:", numeric_df['price'].mean())
print("Maximum Price:", numeric_df['price'].max())
print("Minimum Price:", numeric_df['price'].min())
```
